# Bar Chart Labels with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
np.random.seed(42)
fig, ax = plt.subplots()
# Fake some data
company = ('FB','AMZN','NFLX','GOOG')
y_pos = np.arange(len(company))
performance = 4 + 10 * np.random.rand(len(company))
err = np.random.rand(len(company))

hbars = ax.barh(y_pos, performance, xerr=err, align="center")
ax.set_yticks(y_pos)
ax.set_yticklabels(company)
ax.invert_yaxis() # labels top to bottom
ax.set_xlabel("Performance")
ax.set_title("FAANG Stock Performance Q1")

ax.bar_label(hbars, labels=['Â±%.2f' % e for e in err],
            padding=10, color='r',fontsize=14)
ax.set_xlim(right=18)
plt.show()
```
